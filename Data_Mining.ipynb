```python
import pandas as pd
```

```python
aug = pd.read_csv('./dataset/raw_dataset/russia_ukraine_war_aug2022.csv', lineterminator='\n')
apr = pd.read_csv('./dataset/raw_dataset/russia_ukraine_war_apr2022.csv', lineterminator='\n')
feb = pd.read_csv('./dataset/raw_dataset/russia_ukraine_war_feb2022.csv', lineterminator='\n')
jul = pd.read_csv('./dataset/raw_dataset/russia_ukraine_war_jul2022.csv', lineterminator='\n')
mar = pd.read_csv('./dataset/raw_dataset/russia_ukraine_war_mar2022.csv', lineterminator='\n')
```

```python
combined = pd.DataFrame()

combined = pd.concat([feb, mar, apr, jul, aug], axis=0, ignore_index=True)
combined.drop(['Unnamed: 0'], axis='columns', inplace=True)
combined.head(10)
```

```python
combined.to_csv('./dataset/raw_dataset/raw_combined_dataset.csv')
```
